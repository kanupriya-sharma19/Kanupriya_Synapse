```python
customers=[[5,2],[5,4],[10,3],[20,1]]

current_time = 0
total_time = 0

for arrival_time, prep_time in customers:
    if current_time < arrival_time:
        current_time = arrival_time
    
    waiting_time = current_time + prep_time - arrival_time
    
    total_time =total_time+ waiting_time
    
    current_time =current_time+ prep_time

average_waiting_time = total_time / len(customers)

print("The average waiting time is", average_waiting_time)
```
