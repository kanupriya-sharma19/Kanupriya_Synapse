### The previous commands were just to make sure we have a workable dataframe now, we'll do some manipulations on the data.

Find out the top 5 most voted candidates in Telangana

```python
telangana = df[df['State'] == 'Telangana']
top= telangana.sort_values(by='Total Votes', ascending=False)
top_5= top.head(5)
print("Top 5 most voted candidates in Telangana:")
print(top_5[['State', 'Candidate', 'Total Votes']])
```

Find out the total number of votes received by each state

```python
total_votes_by_state = df.groupby('State')['Total Votes'].sum()
for state, total_votes in total_votes_by_state.items():
    print(f'The total votes of {state} is {total_votes}')
```

Show these values as a percentage of the total number of votes received in the election.

```python
total_votes_by_state = df.groupby('State')['Total Votes'].sum()
for state, total_votes in total_votes_by_state.items():
    percent = (total_votes /df['Total Votes'].sum()) * 100
    print(f'The total votes of {state} is {percent}%')
```

Now, display the top 5 parties and find out the number of votes received by them

```python
total_votes_by_party = df.groupby('Party')['Total Votes'].sum()
top_5_parties = total_votes_by_party.sort_values(ascending=False).head(5)
print("Top 5 most voted parties:")
for party, total_votes in top_5_parties.items():
    print(f'The total votes of {party} is {total_votes}')
```

From the candidates who lost, list 10 candidates who got the highest percent votes.

```python
lost= df[df['Result'] == 'Lost']
lost_top_10 = lost.sort_values(by='Total Votes', ascending=False).head(10)
print(lost_top_10[['State', 'Candidate', 'Total Votes','Result']])
```

### These techniques help to understand the data that we have but we like something a bit more visual right? That's where matplotlib comes to our aid.

Make a piechart to show the votes recieved by the top 5 candidates in the elections.

```python
top=df.sort_values(by='Total Votes', ascending=False)
top_5= top.head(5)
candidates = top_5['Candidate']
votes = top_5['Total Votes']
plt.figure(figsize=(10, 7))
plt.pie(votes, labels=candidates,autopct='%1.1f%%')
plt.title('Votes Received by Top 5 Candidates')
plt.show()
```

Make a bar graph to see the postal votes received by the parties in Karnataka

```python
karnataka = df[df['State'] == 'Karnataka']
#Top as the graph was getting mixed up
top = karnataka.sort_values(by='Postal Votes',ascending=False).head(100)
parties = top ['Party']  
votes = top ['Postal Votes']
plt.figure(figsize=(20, 15))
plt.barh(parties, votes)
plt.title('Top Parties by Postal Votes in Karnataka')
plt.xlabel('Party')
plt.ylabel('Postal Votes')
plt.tight_layout() 
plt.show()
```
