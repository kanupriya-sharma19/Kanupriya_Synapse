```python
request_spending={
    'Mahek':{
        'balance':3000.00,
        'transaction': [
            {"amount":-9000.00,'category':'Creatives'},
            {"amount":7000.00,'category':'Sponsor'},
            {"amount":-2000.00,'category':'Prize-Money'}
        ]
    },
     'Arham':{
        'balance':5000.00,
        'transaction': [
            {"amount":8000.00,'category':'Stalls'},
            {"amount":7500.00,'category':'Seminars'},
           
        ]
    },
     'Unnati':{
        'balance':3500.00,
        'transaction': [
            {"amount":-5000.00,'category':'Attraction'},
            {"amount":2500.00,'category':'Promo'},
            {"amount":-900.00,'category':'Lighting'},
            {"amount":-3000.00,'category':'Games'}
        ]
    },
     'Gaurang':{
        'balance':2000.00,
        'transaction': [
            {"amount":-1500.00,'category':'Website'},
            {"amount":-1000.00,'category':'C2C'},
            {"amount":-500.00,'category':'Posters'}
        ]
    },
    
}
```

```python

def total_spending(request_spending, account_id, category):   
    total_amount=0
    transactions = request_spending[account_id]['transaction']
    for transaction in transactions:
        if(transaction['category'] == category):
             if(transaction['amount']<0):
                total_amount = total_amount-(transaction['amount'] )
    return total_amount

total_amount=total_spending(request_spending, 'Mahek', 'Sponsors')
print('Total amount spent for Sponsor category by Mahesk is' ,total_amount)
total_amount=total_spending(request_spending, 'Arham', 'Stalls')
print('Total amount spent for Stalls category by Arham is' ,total_amount)
total_amount=total_spending(request_spending, 'Unnati', 'Games')
print('Total amount spent for Games category by Unnati is' ,total_amount)
total_amount=total_spending(request_spending, 'Gaurang', 'Website')
print('Total amount spent for Website category by Gaurang is' ,total_amount)
print("===============================================================================")
```

```python

def account_balance(request_spending, account_id):
    total_amount=0
    initial_balance = request_spending[account_id]['balance']
    transactions = request_spending[account_id]['transaction']
    for transaction in transactions:
            total_amount = total_amount+(transaction['amount'] )

    final_balance=initial_balance+total_amount
    return final_balance
  
net_balance=account_balance(request_spending, 'Mahek')
print('Total amount for Mahek is' ,net_balance,'at the end of event')
net_balance=account_balance(request_spending, 'Arham')
print('Total amount for Arham is' ,net_balance,'at the end of event')
net_balance=account_balance(request_spending, 'Unnati')
print('Total amount for Unnati is' ,net_balance,'at the end of event')
net_balance=account_balance(request_spending, 'Gaurang')
print('Total amount for Gaurang is' ,net_balance,'at the end of event')
print("===============================================================================")
```
