```python
from itertools import combinations

Kevin = {'Halsey', 'Taylor Swift', 'Mitski', 'Joji', 'Shawn Mendes', 'Sabrina Carpenter', 'Nicki Minaj', 'Conan Gray', 'One Direction', 'Justin Bieber'}
Stuart = {'Kendrick Lamar', 'Steve Lacy', 'Tyler the Creator', 'Joji', 'The Weeknd', 'Coldplay', 'Kanye West', 'Travis Scott', 'Frank Ocean', 'Brent Faiyaz'}
Bob = {'Conan Gray', 'Joji', 'Dove Cameron', 'Mitski', 'Arctic Monkeys', 'Steve Lacy', 'Kendrick Lamar', 'Isabel LaRosa', 'Shawn Mendes', 'Coldplay', 'Lauv'}
Edith = {'Metallica', 'Billie Eilish', 'The Weeknd', 'Mitski', 'NF', 'Conan Gray', 'Kendrick Lamar', 'Nicki Minaj', 'Kanye West', 'Coldplay'}

DJ_dict = {
    'Kevin': Kevin,
    'Stuart': Stuart,
    'Bob': Bob,
    'Edith': Edith
}

DJ = ['Kevin', 'Stuart', 'Bob', 'Edith']
Combinations = []
#Getting the possible pairs of DJs
c = combinations(DJ, 2)
for combo in c:
    Combinations.append(combo)

finalSingers=[]
djs=[]
invalid=[]
invalid_dj=[]
#Finding the common singers
def FavsOfDJs(PairOfDJs):
    i=0
    print("DJs ------------Common singers")
    for pair in PairOfDJs:
        first_set = DJ_dict[PairOfDJs[i][0]]
        second_set = DJ_dict[PairOfDJs[i][1]]
        fav=[PairOfDJs[i][0],PairOfDJs[i][1]]
        djs.append(fav)
        inter= first_set.intersection(second_set)
        print(PairOfDJs[i][0],'-',PairOfDJs[i][1],':',inter)
        finalSingers.append(inter)
        i+=1
```

```python
#Finding the DJs that do not satisfy the criteria
def InvaildDJs(finalSingers,PairOfDJs,djs):
   
    for f in finalSingers:
        i=0
        first_set = DJ_dict[PairOfDJs[i][0]]
        second_set = DJ_dict[PairOfDJs[i][1]]
        if(len(f)>=len(first_set)*0.3 and len(f)>=len(second_set)*0.3 ):
            pass
        else:
            invalid.append(f)
            invalid_dj.append([PairOfDJs[i][0],PairOfDJs[i][1]])
        i+=1

    for item in invalid:
        if item in finalSingers:
            finalSingers.remove(item)


    for item in invalid_dj:
        if item in djs:
            djs.remove(item)
```

```python

print("Combination of pair of DJs possible and their common singers")
FavsOfDJs(Combinations)
print("===============================================================================")
print("Sorted list of DJs passing the critierias are")
InvaildDJs(finalSingers,Combinations,djs)

#Sorting the DJs based on more number of common singers

sorted_results = sorted(zip(djs, finalSingers), key=lambda x:len(x[1]), reverse=True)
for djs, fav in sorted_results:
        print(djs, fav)
```
