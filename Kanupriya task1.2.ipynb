```python
gadgets=[('Explosive Batarngs',3,True),
        ('Batarngs',5,True), 
        ('Smoke Pellets',0,False),
        ('Tear Gas Grenades',2,True),
        ('Night Vision Googles',1,True),
        ('Batclaw',0,False),
        ('Grapple Gun',3,True),
        ('Batsignal',0,False),
        ('Utility Belt',1,True),
        ('Batmobile Tires',4,True),]

print("Gadgets before sorting")
for item in gadgets:
    print(item)

print("===============================================================================")

sorted_gadgets = sorted(sorted(sorted(gadgets, key=lambda x: x[0]), key=lambda x: x[1],reverse=True), key=lambda x: x[2],reverse=True)
print("Gadgets after sorting")
for item in sorted_gadgets:
    print(item)
```
